```python
# 1)   
#  ()    : z-test()

import numpy as np
import pandas as pd
from scipy.stats import norm

#   
data = np.array([23, 25, 28, 30, 26, 27, 29, 32, 31, 28])

#    ()
population_mean = 28
population_variance = 4

#  
alpha = 0.05

#      ( :  )
alternative = "two.sided"

#  
sample_mean = np.mean(data)

#  
n = len(data)

#   (Z-)
test_statistic = (sample_mean - population_mean) / (np.sqrt(population_variance) / np.sqrt(n))

#   (z-,  )
critical_value = norm.ppf(1 - alpha / 2)

# p-  ( )
p_value = 2 * (1 - norm.cdf(abs(test_statistic)))

#  
print(f" : {sample_mean}")
print(f" : {test_statistic}")
print(f" : {critical_value}")
print(f"p- : {p_value}")

#  
if abs(test_statistic) > critical_value:
    print(" .    .")
else:
    print("  .       .")
```

```python
#   ()   : t-test ()

# stats.ttest 1samp() 
# scipy.stats.ttest_1samp(a, popmean, axis=0, nan_policy='propagate', alternative='two-sided', *, keepdims=False)
# a :    / popmean :      / alternative :    two-sided ( , ), greater(  ), less(  )   
# axis :     (   None ) / nan_policy : NaN   (   propagate )

#=>  : t_statistic :  t- / p_value :  p-(  )

# SciPy   stats.ttest_1samp() t-  ,   t-(one-sample t-test)   .
#  t-             .
# stats.ttest_1samp()   a    popmean  t-  ,  t- p- .



# stats.interval() 
# stats.t.interval(alpha, df, loc=0, scale=1)
# alpha :    0~1   0.95 95%   / df: (degrees of freedom)   1 
# loc :    ,      / scale : ,   

#=>  : low :  (lower bound) / high :  (upper bound)

# SciPy   stats.t.interval() t-    .
# t-                  .  t-  ,     .
# stats.t.interval()      t-     .      .
#   95%      alpha = 0.05, df    loc   , scale     . 
#  95%      .
```

```python
import numpy as np
from scipy import stats

#   
data = np.array([23, 25, 28, 30, 26, 27, 29,  32, 31, 28])

#    
alpha = 0.5

#     
alternative = 'two-sided'  # greater or less 

# t-   #  0 
t_statistic, p_value = stats.ttest_1samp(data, popmean=0, alternative=alternative)

# 95%  
confidence_interval = stats.t.interval(1-alpha, len(data)-1, loc=np.mean(data), scale=stats.sem(data))

#  
print(f"t-statistic :{t_statistic}")
print(f"p-value : {p_value}")
print(f"95% confidence_interval : {confidence_interval}")

#  
if p_value < alpha:
    print(" .    0 .")
else:
    print("  .       .")
```

```python
#  #

# mtcars data     
# 1. mpg    . (   )
# 2. mpg    . (   )
# 3. mpg  20    .
# 4. mpg  95%  .

import numpy as np
import pandas as pd
from scipy import stats

#  
mtcars = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/mtcars.csv')

# mpg     
mpg_mean = np.round(np.mean(mtcars['mpg']), 4)
print(f"mpg  : {mpg_mean}")

# mpg     
mpg_var = np.round(np.var(mtcars['mpg']), 4)
print(f"mpg  : {mpg_var}")

# , t- , mpg   20 
t_stat, p_value = stats.ttest_1samp(mtcars['mpg'], 20)
print("  t-  :")
print(f"t- : {t_stat}")
print(f"p- : {p_value}")

# mpg  95%    
conf_interval = stats.t.interval(0.95, len(mtcars['mpg'])-1, loc=np.mean(mtcars['mpg']), scale=stats.sem(mtcars['mpg']))
print(f"mpg  95%   : {round(conf_interval[0], 4)} - {round(conf_interval[1], 4)}")
```

```python
# 2)   ,   
#        t-  Mann-Whitney U  . t-        , Mann-Whitney U     .

# (1) stats.ttest_ind()
# SciPy  scipy.stats.ttest_ind()     t-(independent samples t-test)   ,          

# scipy.stats.ttest_ind(a, b, axis=0, equal_var=True, nan_policy='propagate')
# a :    , Numpy , Pandas Series, Python        
# b :    , ,,
# axis (optional) :  0,         
# equal_var (optional) :  True,        , True    t- , False       
# nan_policy (optional) :  'propagate' NaN( )   . 'propagate' NaN , 'omit' NaN   

#  p-value        .
```

```python
import numpy as np
from scipy import stats

#    
sample1 = [23, 25, 28, 30, 32]
sample2 = [19, 21, 24, 26, 29]

# t- 
t_statistic, p_value = stats.ttest_ind(sample1, sample2)

#   
mean_sample1 = np.mean(sample1)
mean_sample2 = np.mean(sample2)

#  
alpha = 0.05
print(f"t-statistic : {t_statistic}")
print(f"Sample1 mean : {mean_sample1}")
print(f"Sample2 mean : {mean_sample2}")
print(f"p-value (two_tailed) : {p_value}")
if p_value < alpha:
    print("    .")
else:
    print("     .")
```

```python
# (2) stats.mannwhitneyu() 
# stats.mannwhitneyu()  Mann-Whitney U   Wilcoxon-Mann-Whitney   (non-parametric)       .
#          ,          .

# scipy.stats.mannwhitneyu(x, y, use_continuity=True, alternative=None)
# X & y :    , Numpy , Pandas Series, Python        
# use_continuity (optional) :  True,    (continuity correction)   .       ,    False 
# alternative (optinal) :  ,  two.sided  (    ), less  (      ), greater  (      )

#  p-value         
```

```python
import numpy as np
from scipy import stats

#    
sample1 = [23, 25, 28, 30, 32]
sample2 = [19, 21, 24, 26, 29]

# Mann-Whitney U  
u_statistic, p_value = stats.mannwhitneyu(sample1, sample2)

#   
mean_sample1 = np.mean(sample1)
mean_sample2 = np.mean(sample2)

#  
alpha = 0.05
print(f"u-statistic : {u_statistic}")
print(f"Sample1 mean : {mean_sample1}")
print(f"Sample2 mean : {mean_sample2}")
print(f"p-value (two_tailed) : {p_value}")
if p_value < alpha:
    print("    .")
else:
    print("     .")
```

```python
# (3)   (Normality Test)

#          . 
#         , - , -   ,         .

#     H0 "  ".
# stats.shapiro(X) : X   1       =>   : statistic, p_value
# p-    ,        .
```

```python
from scipy import stats
import numpy as np

#      
data = [2.4, 2.7, 3.1, 3.2, 3.5, 3.7, 3.9, 4.1, 4.2]

# Shapiro-Wilk  
statistic, p_value = stats.shapiro(data)

#  
alpha = 0.05
print(f"Shapiro-Wilk test statistic : {statistic}")
print(f"p-value : {p_value}")
if p_value > alpha:
    print('  ,     .')
else:
    print(' ,      .')
```

```python
#  #

#          .
# (1)    
# (2)      
# (3)   (2)            . ( 0.05)

from scipy import stats
import numpy as np

#  
dat_M = [117, 108, 105, 89, 101, 93, 96, 108, 108, 94, 93, 112, 92, 91, 100, 96, 120, 86, 96, 95]
dat_F = [121, 101, 102, 114, 103, 105, 101, 131, 96, 109, 109, 113, 115, 94, 108, 96, 110, 112, 120, 100]

#  
mean_M = np.mean(dat_M)
mean_F = np.mean(dat_F)
print(f"   : {mean_M}")
print(f"   : {mean_F}")

# -   
statistic_M, p_value_M = stats.shapiro(dat_M)
statistic_F, p_value_F = stats.shapiro(dat_F)

print("\n   Shapiro-Wilk  :", statistic_M)
print(f"  p-value : {p_value_M}")
if p_value_M > 0.05 :
    print("     .")
else:
    print("      .")

print("\n   Shapiro-Wilk  :", statistic_F)
print(f"  p-value : {p_value_F}")
if p_value_F > 0.05 :
    print("     .")
else:
    print("      .")
```

```python
# ,   p-value         .  t- 

#  t- 
t_statistic, p_value = stats.ttest_ind(dat_M, dat_F)

print(f"t-statistic : {t_statistic}")
print(f"p-value : {p_value}")

alpha = 0.05
if p_value < alpha:
    print("      . ( )")
else:
    print("      . ( )")
```
