##### 통계적 가설 검정

1. 가설 검정
- 모집단에 대해 어떤 가설을 설정하고 그 모집단으로부터 추출된 표본을 분석함으로써 그 가설이 틀리는지 맞는지 타당성 여부를 결정(검정)하는 통계적 기법이다.
- 모집단의 특성에 대한 주장은 옳을 수도 있고 틀릴 수도 있다. 객관적이고 과학적인 판단을 위해서는 표본을 선택하여 그 표본을 이용한 결과를 이용하여 가설을 검정해야 한다.
- * 검정통계량(Test Statistic) : 연구자에 의해 설정된 가설은 표본을 근거로 하여 채택여부를 결정짓게 되는데 이때 사용되는 표본통계량을 검정통계량이라 정의한다.
- * 가설 검정(Hypothesis Testing) : 검정통계량의 표본분포에 따라 채택여부를 결정짓는 일련의 통계적 분석과정을 가설 검정이라 하며 일반적으로 몇 단계의 절차를 거쳐 검정이 수행된다.

---

- (1) 가설 검정의 절차
* ① 가설의 설정 : 집단의 특성을 파악하기 위해서 표본을 이용한 의사결정은 오류의 가능성이 상존한다. 따라서 가설 검정은 오류의 가능성을 사전에 관리하는 것이 중요하다. 
**오류의 허용확률을 정해 놓고 그 기준에 따라 가설의 채택이나 기각을 결정**한다.
- * 귀무가설(Null Hypothesis, H0) : 현재 통념적으로 믿어지고 있는 모수에 대한 주장 또는 원래의 기준이 되는 가설이다.
- * 대립가설(Alternative Hypothesis, H1) : 연구자가 모수에 대해 새로운 통계적 입증을 이루어 내고자 하는 가설이다.
- 표본을 통해 새롭게 주장하는 대립가설이 충분히 입증되지 못한다면, 연구자는 현재 믿어지고 있는 주장인 귀무가설을 그대로 받아들여야 할 것이다.

* ② 유의수준 α
- 가설 검정의 결과로 가설의 채택여부를 결정하게 될 때 우리는 두 가지의 오류를 생각할 수 있다.

- * 제 1종 오류(Type Ⅰ종 Error) : 귀무가설이 참일 때 귀무가설을 기각하도록 결정하는 오류 (즉, 대립가설을 채택, 무죄인데 유죄라고 할 오류 - 더 중요함)

- * 제 2종 오류(Type Ⅱ종 Error) : 귀무가설이 거짓인데 귀무가설을 채택할 오류, 또는 대립가설이 참일 때 귀무가설을 채택하도록 결정하는 오류 (즉, 대립가설을 기각, 유죄인데 무죄라고 할 오류)

- 유의수준(Significance Level) : 제 1종 오류를 범할 확률의 최대 허용한계, 즉 표본에서 얻은 표본통계량이 기각역(regjection region)에 들어갈 확률을 유의수준 또는 위험률(risk ratio)이라고 하며 가설 검정에서 판단의 기준으로 삼고 있다.
- => 유의수준 α는 귀무가설이 참인데도 이것을 기각하게 될 확률을 말한다. 일반적으로 1%, 5%, 10% 유의수준 등이 많이 이용된다.
- => 귀무가설이 맞는데 틀렸다고 결론 내렸을 확률(귀무가설을 잘못 기각할 확률), 즉 유의수준이 낮을수록 연구자는 귀무가설을 기각하고 자신의 주장에 확실을 가질 수 있다.

- * p - value(유의 확률) : 관찰된 데이터의 검정통계량이 귀무가설을 지지하는 정도를 확률로 표현한 것이다.
- => p-value가 0.07인 경우 귀무가설을 기각하면 잘못 기각할 확률이 0.07이 되고, 0.03인 경우 귀무가설을 기각하면 잘못 기각할 확률이 0.03이 된다. 따라서 전자의 경우보다 후자의 경우 귀무가설을 보다 자신 있게 기각하는 것이다.

- * 귀무가설의 기각여부는 p-value와 α의 크기에 달려 있다. 즉, p-value가 작을수록 그리고 α의 값이 클수록 귀무가설을 기가할 수 있다.
- * 유의수준은 곧 1종 오류의 확률이다. 허용 유의수준 α값은 보통 0.05로 정해지지만 경우에 따라서는 0.01 혹은 0.1이 사용되기도 한다. 0.05라는 것은 95% 신뢰수준을 의미한다.

---

- (2) 검정 통계량 및 표본분포의 결정
- 모수에 대한 정보는 표본에 함축되어 있다. 따라서 표본을 통하여 가설의 채택 여부를 결정짓게 되는데, 이때 사용되는 표본 통계량을 검정통계량이라 한다. 또한 유의수준에 따른 귀무가설의 기각역을 결정하기 위해서, 귀무가설이 참일 때 검정통계량의 확률분포를 알아야만 하며, 알려져 있지 않을 때에는 통계학의 극한이론에 근거하여 근사적인 분포가 정해져야 한다.

---

- (3) 기각역의 결정
- 표본에서 계산된 통계량이 가설로 설정한 모집단의 성격과 현저한 차이가 있을 경우에는 모집단에 대해 설정한 귀무가설을 기각하게 된다.
- * 이때 귀무가설을 기각하게 되는 검정통계량의 범위를 기각역(Critical Region, Rejection Region)이라 하며, 기각역의 경계값을 임계치라 한다.
- => 임계치(Critical Value) : 주어진 유의수준 α에서 귀무가설의 채택과 기각에 관련된 의사결정을 할 때, 그 기준이 되는 점이다.
- * 기각역은 검정통계량의 확률분포(귀무가설이 참일 때)와 유의수준 α에서 대립가설의 형태(우측, 좌측 또는 양측)에 따라 단측 또는 양측 검정통계량이 설정된다.
- * 양측 검정 : 가설 검정에서 기각영역이 양쪽에 있는 것이다.
- * 단측 검정 : 가설 검정에서 기각영역이 어느 한쪽에만 있게 되는 경우이다.

---

- (4) 검정통계량의 계산
- ① 의사결정
- 표본의 관측치로부터 계산된 검정통계량의 값이 기각역에 속하면 귀무가설을 기각하며(대립가설을 채택) 그렇지 않으면 귀무가설을 채택(대립가설을 기각)한다.
- ② 통계량의 계산과 임계치 비교
- 임계치가 결정되면 표보에서 얻은 통계량이 기각영역에 속하는지 채택 영역에 속하는지를 결정해야 한다.
- 임계치는 x̄, Z, t값으로 나타낼 수 있다. 예를 들어 표본을 기초로 계산된 Z(t)값을 계산된 Z(t) 값 (computed Z(t) - value)라고 부른다.
- 계산한 값을 임계치와 비교해서 기각영역 안에 있으면 H0을 기각하고 채택 영역 안에 있으면 H0를 채택한다.
- ③ p-value
- 주어진 자료로서 귀무가설을 기각하려고 할 때 필요한 최소의 유의수준을 의미하며, 다른 용어로 유의성 확률(Significant Probability) 또는 관측된 유의수준(Observed Significance Level)이라고도 한다. p-value가 계산되는 경우에는 유의수준 α와 비교하여 다음과 같은 결정을 할 수 있다.
- => p-value < α : 귀무가설을 기각
- => p-value > α : 귀무가설을 채택

```python
# 1) 표본의 평균 검정
# 모집단의 분산(표준편차)을 알고 있는 경우 : z-test(양측)

import numpy as np
import pandas as pd
from scipy.stats import norm

# 예제 데이터 생성
data = np.array([23, 25, 28, 30, 26, 27, 29, 32, 31, 28])

# 모집단의 평균과 분산 (표준편차)
population_mean = 28
population_variance = 4

# 유의수준 설정
alpha = 0.05

# 단측 또는 양측 검정 선택 (예 : 양측 검정)
alternative = "two.sided"

# 표본평균 계산
sample_mean = np.mean(data)

# 표본 크기
n = len(data)

# 검정통계량 계산 (Z-검정)
test_statistic = (sample_mean - population_mean) / (np.sqrt(population_variance) / np.sqrt(n))

# 임계값 계산 (z-검정, 양측 검정)
critical_value = norm.ppf(1 - alpha / 2)

# p-값 계산 (양측 검정)
p_value = 2 * (1 - norm.cdf(abs(test_statistic)))

# 결과 출력
print(f"표본평균 : {sample_mean}")
print(f"검정통계량 : {test_statistic}")
print(f"임계값 : {critical_value}")
print(f"p-값 : {p_value}")

# 결과 해석
if abs(test_statistic) > critical_value:
    print("귀무가설을 기각합니다. 표본평균이 모집단 평균과 다릅니다.")
else:
    print("귀무가설을 기각하지 못합니다. 표본평균이 모집단 평균과 차이가 없을 가능성이 있습니다.")
```

```python
#  모집단의 분산(표준편차)을 모르는 경우 : t-test (양측)

# stats.ttest 1samp() 함수
# scipy.stats.ttest_1samp(a, popmean, axis=0, nan_policy='propagate', alternative='two-sided', *, keepdims=False)
# a : 검정하려는 데이터 배열 / popmean : 비교하려는 가설값 또는 모집단의 평균값 / alternative : 대립가설을 지정하는 옵션으로 two-sided (양측 검정, 기본값), greater(우측 단측 검정), less(좌측 단측 검정) 중 하나를 선택
# axis : 계산을 수행할 축을 지정 (일반적으로 디폴트 값인 None을 사용) / nan_policy : NaN값의 처리 방법을 지정(일반적으로 디폴트 값인 propagate를 사용)

#=> 반환값 : t_statistic : 계산된 t-통계량 / p_value : 계산된 p-값(검정 결과의 유의확률)

# SciPy 라이브러리에서 제공하는 stats.ttest_1samp()함수는 t-검정을 수행하는데 사용되며, 단일 표본 t-검정(one-sample t-test)를 수행하는 데 유용하다.
# 단일표본 t-검정은 특정 모집단의 평균값이 어떤 가설 값과 일치하는지를 확인하는 통계적 검정 방법 중 하나이다.
# stats.ttest_1samp() 함수는 데이터 a를 가지고 모집단의 평균값 popmean에 대한 t-검정을 수행하여 , 결과인 t-통계량과 p-값을 반환한다.



# stats.interval() 함수
# stats.t.interval(alpha, df, loc=0, scale=1)
# alpha : 신뢰수준을 나타내며 주로 0~1사이의 값으로 예를들어 0.95는 95%의 신뢰수준을 나타냄 / df: 자유도(degrees of freedom) 표본 크기에서 1을 뺀값
# loc : 신뢰구간의 중심이 되는 값, 일반적으로 표본 데이터의 평균을 사용 / scale : 표준오차, 평균의 표준오차를 사용

#=> 반환값 : low : 신뢰구간의 하한(lower bound) / high : 신뢰구간의 상한(upper bound)

# SciPy 라이브러리에서 제공하는 stats.t.interval()함수는 t-분포를 사용하여 신뢰구간을 게산하는데 사용된다.
# t-분포를 사용하는 이유는 표본 크기가 작거나 모집단의 분산이 알려져 있지 않을 때 평균에 대한 신뢰구간을 계산하는 데 유용하기 때문이다. 주로 t-검정과 함께 사용되며, 데이터의 평균값에 대한 신뢰구간을 계산한다.
# stats.t.interval() 함수는 주어진 신뢰구간과 자유도에 따라 t-분포에 해당 구간의 하한과 상한을 계산한다. 이것은 표본 평균의 신뢰구간을 제공하는데 사용된다.
# 예를 들어 95% 신뢰수준에서 평균값에 대한 신뢰구간을 계산하려면 alpha = 0.05, df를 적절한 값으로 설정하고 loc에 표본 데이터의 평균, scale에 표본 데이터의 표준오차를 전달하면 된다. 
# 결과로는 95% 신뢰수준에서 평균값에 대한 하한과 상한이 반환된다.
```

```python
import numpy as np
from scipy import stats

# 예제 데이터 생성
data = np.array([23, 25, 28, 30, 26, 27, 29,  32, 31, 28])

# 유의 수준 설정 
alpha = 0.5

# 단측 또는 양측 검정 선택
alternative = 'two-sided'  # greater or less로 변경가능

# t-검정 수행  # 모집단평균 0으로 설정
t_statistic, p_value = stats.ttest_1samp(data, popmean=0, alternative=alternative)

# 95% 신뢰구간 계산
confidence_interval = stats.t.interval(1-alpha, len(data)-1, loc=np.mean(data), scale=stats.sem(data))

# 결과 출력
print(f"t-statistic :{t_statistic}")
print(f"p-value : {p_value}")
print(f"95% confidence_interval : {confidence_interval}")

# 결과 해석
if p_value < alpha:
    print("귀무가설을 기각합니다. 표본평균이 모집단 평균 0과 다릅니다.")
else:
    print("귀무가설을 기각하지 못합니다. 표본평균이 모집단 평균과 차이가 없을 가능성이 있습니다.")
```

```python
# 결과 설명 #

# t-statistic(t-검정통계량) : 31.887 이 값은 표본 데이터의 평균이 모집단의 평균과 얼마나 다른지를 나타내는 지표이다. 큰t값은 표본평균과 모집단 평균의 차이가 통계적으로 유의하다는 것을 나타낸다.
# p-value : 1.441e-10으로 매우 작으며 귀무가설을 기각하기에 충분히 낮은 값이다. 따라서 대립가설을 채택한다. 즉, 표본평균이 모집단 평균과 다르다고 할 수 있다.
# 95% confidence interval : 95% 신뢰구간은 (25.9207, 29.8793)이 산출된다. 이 구간은 모집단의 평균에 대한 95% 신뢰 구간을 나타내며, 이 범위 안에 모집단 평균의 진짜 값이 포함될 확률이 95%이다.
```

```python
# 따라하기 #

# mtcars data를 이용하여 다음 아래 물음에 답하시오
# 1. mpg 데이터에 대한 표본평균을 구하시오. (반올림하여 소수 넷째자리까지 계산)
# 2. mpg 데이터에 대한 표본분산을 구하시오. (반올림하여 소수 넷째자리까지 계산)
# 3. mpg의 평균이 20이라는 가설에 대해서 검정을 하시오.
# 4. mpg의 평균의 95% 신뢰구간을 구하시오.

import numpy as np
import pandas as pd
from scipy import stats

# 데이터 불러오기
mtcars = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/mtcars.csv')

# mpg 열의 표본평균 계산 및 반올림
mpg_mean = np.round(np.mean(mtcars['mpg']), 4)
print(f"mpg의 표본평균 : {mpg_mean}")

# mpg 열의 표본분산 계산 및 반올림
mpg_var = np.round(np.var(mtcars['mpg']), 4)
print(f"mpg의 표본분산 : {mpg_var}")

# 단일표본, t-검정 수행, mpg 열의 평균이 20인지 검정
t_stat, p_value = stats.ttest_1samp(mtcars['mpg'], 20)
print("단일 표본 t-검정 결과 :")
print(f"t-통계량 : {t_stat}")
print(f"p-값 : {p_value}")

# mpg 평균의 95% 신뢰구간 계산 및 출력
conf_interval = stats.t.interval(0.95, len(mtcars['mpg'])-1, loc=np.mean(mtcars['mpg']), scale=stats.sem(mtcars['mpg']))
print(f"mpg 평균의 95% 신뢰 구간 : {round(conf_interval[0], 4)} - {round(conf_interval[1], 4)}")
```

```python
# 2) 두 독립표본의 평균, 중앙값 차이 검정
# 두 독립표본의 평균 차이를 검정하기 위해 일반적으로 t-검정 또는 Mann-Whitney U 검정을 사용한다. t-검정은 두 그룹의 데이터가 정규 분포를 따를 때 사용하며, Mann-Whitney U검정은 정규분포를 가정하지 않을 때 사용한다.

# (1) stats.ttest_ind()
# SciPy 라이브러리의 scipy.stats.ttest_ind()함수는 두 독립적인 표본 간의 t-검정(independent samples t-test)을 수행하기 위한 함수로, 두 표본의 평균이 통계적으로 유의미하게 다른지 여부를 검정하는 데 사용

# scipy.stats.ttest_ind(a, b, axis=0, equal_var=True, nan_policy='propagate')
# a : 첫 번째 독립 표본, Numpy 배열, Pandas Series, Python 리스트 또는 다른 유사한 데이터 구조 사용 가능
# b : 두 번째 도릭 표본, ,,
# axis (optional) : 기본값은 0, 표본 간의 평균을 비교할 축을 지정하며 다차원 배열에서 유용
# equal_var (optional) : 기본값은 True, 두 그룹의 분산이 동일한지 여부를 나타내는 부울 값, True로 설정하면 등분산을 가정하고 t-검정을 수행하며, False로 설정하면 등분산 가정을 하지 않는 검정을 수행
# nan_policy (optional) : 기본값은 'propagate' NaN(누락된 값)을 처리하는 방법을 지정. 'propagate'는 NaN을 유지하고, 'omit'은 NaN 값을 무시하고 분석

# 일반적으로 p-value가 유의수준보다 작으면 두 그룹의 평균이 통계적으로 다르다고 판단한다.
```

```python
import numpy as np
from scipy import stats

# 두 독립 표본 데이터
sample1 = [23, 25, 28, 30, 32]
sample2 = [19, 21, 24, 26, 29]

# t-검정 실행
t_statistic, p_value = stats.ttest_ind(sample1, sample2)

# 각 샘플의 평균
mean_sample1 = np.mean(sample1)
mean_sample2 = np.mean(sample2)

# 결과 출력
alpha = 0.05
print(f"t-statistic : {t_statistic}")
print(f"Sample1 mean : {mean_sample1}")
print(f"Sample2 mean : {mean_sample2}")
print(f"p-value (two_tailed) : {p_value}")
if p_value < alpha:
    print("두 그룹의 평균은 유의미하게 다릅니다.")
else:
    print("두 그룹의 평균은 유의미하게 다르지 않습니다.")
```

```python
# (2) stats.mannwhitneyu() 함수
# stats.mannwhitneyu() 함수는 Mann-Whitney U 검정 또는 Wilcoxon-Mann-Whitney 검정으로도 알려진 비모수적인(non-parametric) 두 독립표본 간의 차이를 검정하는 데 사용된다.
# 이 검정을 두 그룹의 중앙값이 서로 다른 지 여부를 평가하며, 정규분포를 따르지 않는 데이터 또는 등분산성을 가정하기 어려운 데이터에 적합하다.

# scipy.stats.mannwhitneyu(x, y, use_continuity=True, alternative=None)
# X & y : 비교하려는 두 독립표본 데이터, Numpy 배열, Pandas Series, Python 리스트 또는 다른 유사한 데이터 구조 사용 가능
# use_continuity (optional) : 기본값은 True, 이 값은 연속성 보정(continuity correction)을 사용할지 여부를 제어. 연속성 보장은 이산적인 데이터에서 사용될 수 있으며, 중요하지 않은 경우 False로 설정
# alternative (optinal) : 대립가설을 정의하며, 기본값 two.sided는 양측 검정(두 그룹의 중앙값이 다르다고 주장), less는 좌측 검정(첫 번째 그룹의 중앙값이 더 낮다고 주장), greater는 우측 검정(첫 번째 그룹의 중앙값이 더 높다고 주장)

# 일반적으로 p-value가 유의수준보다 작으면 두 그룹의 중앙값이 통계적으로 유의미하게 다르다고 판단
```

```python
import numpy as np
from scipy import stats

# 두 독립 표본 데이터
sample1 = [23, 25, 28, 30, 32]
sample2 = [19, 21, 24, 26, 29]

# Mann-Whitney U 검정 실행
u_statistic, p_value = stats.mannwhitneyu(sample1, sample2)

# 각 샘플의 평균
mean_sample1 = np.mean(sample1)
mean_sample2 = np.mean(sample2)

# 결과 출력
alpha = 0.05
print(f"u-statistic : {u_statistic}")
print(f"Sample1 mean : {mean_sample1}")
print(f"Sample2 mean : {mean_sample2}")
print(f"p-value (two_tailed) : {p_value}")
if p_value < alpha:
    print("두 그룹의 중앙값은 유의미하게 다릅니다.")
else:
    print("두 그룹의 중앙값은 유의미하게 다르지 않습니다.")
```

```python
# (3) 정규성 검정 (Normality Test)

# 정규성 검정을 통계 분석에서 데이터가 정규분포를 따르는 지 확인하는 테스트다. 
# 가장 일반적으로 사용되는 정규성 검정으로 샤피로 윌크 정규성 검정, 콜모고로프-스미르노프 검정, 다고스티노-피어슨 검정 등이 있으며, 데이터의 특성을 이해하고 적절한 통계 기법을 선택하는 판단을 돕는다.

# 샤피로 윌크 정규성 검정의 귀무가설H0은 "데이터는 정규분포를 따른다"이다.
# stats.shapiro(X) : X 정규성을 검정하려는 1차원 배열 혹은 리스트 형태의 데이터 샘플 => 반환 값 : statistic, p_value
# p-값이 유의수준보다 작으면 귀무가설 기각하며, 데이터가 정규분포를 따르지 않을 가능성이 높다는 것을 의미한다.
```

```python
from scipy import stats
import numpy as np

# 정규성 검정을 위한 샘플 데이터 생성
data = [2.4, 2.7, 3.1, 3.2, 3.5, 3.7, 3.9, 4.1, 4.2]

# Shapiro-Wilk 테스트 수행
statistic, p_value = stats.shapiro(data)

# 결과 출력
alpha = 0.05
print(f"Shapiro-Wilk test statistic : {statistic}")
print(f"p-value : {p_value}")
if p_value > alpha:
    print('귀무가설을 기각하지 못하며, 데이터는 정규분포를 따를 가능성이 높습니다.')
else:
    print('귀무가설을 기각하며, 데이터는 정규분포를 따르지 않을 가능성이 있습니다.')
```

```python
# 따라하기 #

# 다음 데이터는 초등학교 취학 전 남자와 여자의 키를 추출한 데이터이다.
# (1) 각각의 샘플의 평균을 구하시오
# (2) 각 샘플에 대해서 정규성 테스트를 하시오
# (3) 샘플에 대한 (2)의 결과를 바탕으로 적절한 검정 방법을 선택하여 평균에 대한 차이가 있는지 검정을 수행하시오. (유의수준은 0.05)

from scipy import stats
import numpy as np

# 데이터 샘플
dat_M = [117, 108, 105, 89, 101, 93, 96, 108, 108, 94, 93, 112, 92, 91, 100, 96, 120, 86, 96, 95]
dat_F = [121, 101, 102, 114, 103, 105, 101, 131, 96, 109, 109, 113, 115, 94, 108, 96, 110, 112, 120, 100]

# 평균 계산
mean_M = np.mean(dat_M)
mean_F = np.mean(dat_F)
print(f"남자 데이터 평균 : {mean_M}")
print(f"여자 데이터 평균 : {mean_F}")

# 샤피로-윌크 정규성 테스트 수행
statistic_M, p_value_M = stats.shapiro(dat_M)
statistic_F, p_value_F = stats.shapiro(dat_F)

print("\n 남자 데이터 Shapiro-Wilk 테스트 통계량:", statistic_M)
print(f"남자 데이터 p-value : {p_value_M}")
if p_value_M > 0.05 :
    print("남자 데이터는 정규분포를 따를 가능성이 높습니다.")
else:
    print("남자 데이터는 정규분포를 따르지 않을 가능성이 있습니다.")

print("\n 여자 데이터 Shapiro-Wilk 테스트 통계량:", statistic_F)
print(f"여자 데이터 p-value : {p_value_F}")
if p_value_F > 0.05 :
    print("여자 데이터는 정규분포를 따를 가능성이 높습니다.")
else:
    print("여자 데이터는 정규분포를 따르지 않을 가능성이 있습니다.")
```

```python
# 남자, 여자 모두 p-value가 유의수준과 비교하여 충분히 크므로 정규 분포를 따를 가능성이 높다. 따라서 t-검정 수행

# 독립표본 t-검정 수행
t_statistic, p_value = stats.ttest_ind(dat_M, dat_F)

print(f"t-statistic : {t_statistic}")
print(f"p-value : {p_value}")

alpha = 0.05
if p_value < alpha:
    print("두 표본 간에는 유의한 평균 차이가 있다. (귀무가설 기각)")
else:
    print("두 표본 간에는 유의한 평균 차이가 없다. (귀무가설 채택)")
```

```python
# 3) 대응표본의 평균 차이 검정 (t-검정)
# stats.ttest_rel(a, b, nan_policy='propagate', alternative='two-sided', *, keepdims=False)

# 대응표본 (two related samples) 에 대한 t-검정을 수행하기 위해 사용한다. 두 관련된 데이터 집단 간의 평균 차이를 비교하고, 이 차이가 통계적으로 유의미한 지 확인하는 데 사용

from scipy import stats

data_before = [72, 75, 68, 71, 73]
data_after = [76, 79, 74, 78, 80]

t_statistic, p_value = stats.ttest_rel(data_before, data_after)

print("대응표본 t-검정 결과:")
print(f"t-통계량 : {t_statistic}")
print(f"p-값 : {p_value}")
```
